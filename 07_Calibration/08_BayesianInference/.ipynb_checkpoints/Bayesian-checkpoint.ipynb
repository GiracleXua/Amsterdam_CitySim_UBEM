```python
def detectOutlier(filePath, year, remove = True):
    path = filePath + '/output' + str(year) + '.csv'
    df = pd.read_csv(path, delimiter=',', index_col=False)
    # allow metered data to have 1.5 kWh/m3 deviation threshold to simulation data
    threshold = 1.5
    outlierID = []
    for i in range(len(df)):
        pc6consumption = df.iloc[i, 3]
        simMax = max(df.iloc[i, 4:]) + threshold
        simMin = min(df.iloc[i, 4:]) - threshold
        if not simMin <= pc6consumption <= simMax:
            outlierID.append(i)
            # print str(df.iloc[i, 1]) + ' is an outlier with pc6 consumption = ' + str(df.iloc[i, 3]) + ' kWh/m3_yr'
    if bool(remove):
        df.drop(outlierID, inplace=True)
        cleaned_df = df.reset_index(drop=True).iloc[:,1:]
    return cleaned_df

def dinstinctArchetype(cleaned_df):
    archetypes = cleaned_df.archetype.unique()
    return archetypes

def getArchetypeIDs(cleaned_df, archetype):
    archIDs = cleaned_df.index[cleaned_df['archetype'] == archetype].tolist()
    return archIDs 

def getGroupStatistics(cleaned_df, archetype):
    archIDs = getArchetypeIDs(cleaned_df, archetype)
    mu = np.mean(cleaned_df.loc[archIDs, 'l_pc6_consumption_kwh_m3'])
    std = np.std(cleaned_df.loc[archIDs, 'l_pc6_consumption_kwh_m3'])
    return mu, std

def pc6Metadata(cleaned_df, pc6_name):
    ID = cleaned_df.index[cleaned_df['postcode'] == pc6_name]
    if len(ID) >= 1:
        pc6_ID = int(ID[0])
        archetype = cleaned_df['archetype'][pc6_ID]
    else:
        pc6_ID = 'nan'
        archetype = 'nan'
    return pc6_ID, archetype


# HERE IS BAYESIAN FRAMEWORK PART #

# Likelihood now is based on assuming EUI of specific archetype is a normal PDF
# Given simulation result, the function returns likelihood
# Doing this assumption is based on that archetype itself is already a coarse classification based on EUI
def computelikelihood(meteredConsumption, mu, std):
    # mu: simulation result
    # std: user specified
    likelihood = norm.pdf(meteredConsumption, mu, std)
    return likelihood

# Compute posterior for each postcode
# def computeLogPosterior(cleaned_df, archetype, pc6_index, prior):
#     mu, std = getGroupStatistics(cleaned_df, archetype)
    
#     # get pc6 consumption for this specific postcode
#     pc6_consumption = cleaned_df.loc[pc6_index, 'l_pc6_consumption_kwh_m3']
#     likelihoodXprior = []
    
#     for n in range(25):
#         n += 1
#         nSimConsumption = cleaned_df.loc[pc6_index, 'case'+str(n)+'_kwh_m3']
#         nLikelihood = computelikelihood(nSimConsumption, pc6_consumption, std)
#         # to avoid numerical problem, apply the principle: log(product(Pi))=sum(log(Pi))
#         likelihoodXprior.append(np.log(nLikelihood) + np.log(prior[n-1]))
    
#     # anti-logarithm
#     likelihoodXprior = np.exp(likelihoodXprior)
#     probD = sum(likelihoodXprior)
#     pc6_posterior = np.multiply(likelihoodXprior,100)/probD
#     return pc6_posterior

def computePosterior(cleaned_df, archetype, pc6_index, prior):
    gridSize = 5
    std = 3.5
    
    # get pc6 consumption for this specific postcode
    pc6_consumption = cleaned_df.loc[pc6_index, 'l_pc6_consumption_kwh_m3']
    likelihoodXprior = []
    
    for n in range(gridSize*gridSize):
        nSimConsumption = cleaned_df.loc[pc6_index, 'case'+str(n+1)+'_kwh_m3']
        nLikelihood = computelikelihood(pc6_consumption, nSimConsumption, std)
        #likelihoodXprior.append(nLikelihood*prior[n])
        # to avoid numerical problem, apply the principle: log(product(Pi))=sum(log(Pi))
        likelihoodXprior.append(np.log(nLikelihood) + np.log(prior[n]))
        
    likelihoodXprior = np.exp(likelihoodXprior)
    probD = sum(likelihoodXprior)
    pc6_posterior = np.multiply(likelihoodXprior,100)/probD
    return pc6_posterior

def computeMarginalProb(posterior):
    npMatrix = np.reshape(posterior, (5,5))
    marginalProb = contingency.margins(npMatrix)
    Tmin_Prob = marginalProb[0].transpose()[0]
    Ninf_Prob = marginalProb[1][0]
    return Tmin_Prob, Ninf_Prob


# HERE IS RESULT VISUALIZATION #

def plotHeatMaps(posterior):
    plt.figure(figsize=(8,7))
    
    posteriorMatrix = np.reshape(posterior, (5,5))
    posteriorDF = pd.DataFrame(data=posteriorMatrix, 
                               index = ['15', '16.25', '17.5', '18.75', '20'], 
                               columns=['0.19', '0.345', '0.5', '0.655', '0.81'])
    #print posteriorDF
    sns.heatmap(posteriorDF, cmap="Blues", alpha=0.7, annot=True, annot_kws={"size": 8})
    
    plt.tick_params(axis='both', labelsize=13)
    plt.xlabel('Infiltration rate', fontsize=14)
    plt.ylabel('Temperature', fontsize=14)
    return plt
    
def plotMarginalPDF(Prob, paraName):
    if paraName == 'p_Ninf':
        bars = ('0.19', '0.345', '0.5', '0.655', '0.81')
        # xLabel = 'Infiltration rate'
    elif paraName == 'p_Tmin':
        bars = ('15', '16.25', '17.5', '18.75', '20')
        # xLabel = 'Temperature'
        
    palette = plt.get_cmap('Set1')
    y_pos = np.arange(len(bars))
    plt.bar(y_pos, Prob, color=palette(1), alpha=0.7)
    plt.xticks(y_pos, bars)
    xLabel = 'Top: Temperature, Bottom: Infiltration rate'
    plt.tick_params(axis='both', labelsize=12)
    plt.xlabel(xLabel, fontsize=12)
    plt.ylabel("Probability (%)", fontsize=12)
    
def plotDiagrams(posterior, pc6_name, yr):
    heatmap = plotHeatMaps(posterior)
    heatmap.savefig(filePath+'/plots/'+str(pc6_name)+'_HeatMap_year'+str(yr))
    heatmap.show()
    
    Tmin_Prob, Ninf_Prob = computeMarginalProb(posterior)
    plt.subplot(2,1,1)
    plotMarginalPDF(Tmin_Prob, 'p_Tmin')
    plt.subplot(2,1,2)
    plotMarginalPDF(Ninf_Prob, 'p_Ninf')
    plt.savefig(filePath+'/plots/'+str(pc6_name)+'_year'+str(yr))
    plt.show()
    
```

```python
def trainingPhase(years, pc6_name):    
    nTrainingYear = 1
    gridSize = 5
    
    jointProb = []
    for yr in years:
        cleaned_df = detectOutlier(filePath, yr, True)
        pc6_ID, archetype = pc6Metadata(cleaned_df, pc6_name)
        if type(pc6_ID) == int:
            print 'year: ', yr
            print cleaned_df.iloc[pc6_ID, 0:3]
            
            # initialize prior to be equal distribution
            if nTrainingYear == 1:
                prior = [(1.0/gridSize)**2]*(gridSize*gridSize)
            # posterior become a new prior in the subsequent years
            else:
                prior = posterior
            
            posterior = computePosterior(cleaned_df, archetype, pc6_ID, prior)
            plotDiagrams(posterior, pc6_name, yr)
            nTrainingYear += 1
            jointProb.append(posterior)
            #jointProb = np.multiply(jointProb, posterior)
        else: 
            print 'No usable ' + str(pc6_name) + ' consumption data available at year: ', yr
            pass
    return jointProb
        
        
years = ['2010','2011','2012','2013', '2014', '2015']
jointProb = trainingPhase(years, '1094SH')
```

```python
def computeJointProb(jointProb):
    # To avoid numerical prob, that is, to avoid 0 probability exist
    cleanedJoint = []
    for yearPost in jointProb:
        n = 0
        for prob in yearPost:
            if prob <= 1e-16:
                yearPost[n] = 1e-16
            n += 1
        pDenominator = sum(yearPost)
        yearPost = np.multiply(yearPost, 100)/pDenominator
        cleanedJoint.append(yearPost)
    return cleanedJoint
    
#cleanedJoint = computeJointProb(jointProb)
pD = sum(np.e**sum(np.log(jointProb)))
finalJoint = np.multiply(100,np.e**sum(np.log(jointProb)))/pD

plotDiagrams(finalJoint,'non','non')
```
